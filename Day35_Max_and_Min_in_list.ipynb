```python
A=int(input("Enter how many numbers you want to enter"))
numbers=[]

for i in range(A):
   num=int(input(f"Enter number{i+1}:"))
   numbers.append(num)

max_num=numbers[0]
min_num=numbers[0]

for num in numbers:
   if num>max_num:
     max_num=num
   if num<min_num:
     min_num=num

print("maximum number:",max_num)
print("minimum number:",min_num)
```
