```python
A=input("Enter positive number:")
reverse_A=A[::-1]
sum_digits=0

for ch in A:
   digit=int(ch)
   sum_digits=digit+sum_digits

print("Reversed number is:",reverse_A)
print("Sum of digit is:",sum_digits)
```

Here,we had first enter the positive number as strings...for reverse.Then using loop,we had changed each character as digit and sum one by one.

```python
s = input("Enter an integer: ")  # can be negative too

# Step 1: Check if it's negative
if s[0] == '-':
    is_negative = True
    s = s[1:]          # remove the '-' sign temporarily
else:
    is_negative = False

# Step 2: Reverse the digits
reversed_s = s[::-1]

# Step 3: Find sum of digits
sum_digits = 0
for ch in s:
    sum_digits += int(ch)

# Step 4: Add back '-' sign if needed
if is_negative:
    reversed_s = '-' + reversed_s

# Step 5: Show results
print("Reversed number:", reversed_s)
print("Sum of digits:", sum_digits)
```
