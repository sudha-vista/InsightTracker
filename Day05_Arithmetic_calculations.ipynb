```python
A=float(input("Enter the first number:"))
B=float(input("Enter the second number:"))

sum=A+B
difference=A-B
division=A/B
product=A*B
remainder=A%B
power=A**B
floordivision=A//B

print("Enter sum of the two numbers:",sum)
print("Enter difference of two numbers:",difference)
print("Enter the division of two numbers:",division)
print("Enter the product of two numbers:",product)
print("Enter the remainder of two numbers:",remainder)
print("Enter the power of two numbers:",power)
print("Enter the floor division of two numbers:",floordivision)
```

#program- Perform all Arithmetic operations






**Notes**

* floor division(//)- divides but ignore decimal parts
* modulus(%)- Gives the remainder after division
* exponent/power(**)- raises to power
