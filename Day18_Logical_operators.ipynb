```python
A=20
B=100
C=30

L= A>B and A<B
M= A<B or B<C
N= not(A>B)

print("A>B and A<B:",L)
print("A<B or B<C:",M)
print("not(A>B):",N)
```

NOTES:

✨ Operators:

and → True if both conditions are True

or → True if any one condition is True

not → Reverses the condition (True → False, False → True)
