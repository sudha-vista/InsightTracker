```python
a=10
b=20
print("Before swapping:")
print("A=",a)
print("B=",b)

temp=a #(temp temporarily stores a's value(10)) - using third variable
a=b
b=temp

print("After swapping:")
print("A=",a)
print("B=",b)
```

```python
a=10
b=20
print("Before swapping:")
print("A=",a)
print("B=",b)

a,b=b,a #(swap using python shortcut) - without using third varaible

print("After swapping:")
print("A=",a)
print("B=",b)
```

Swap two numbers in python

* Program1- Using a Third Variable
* Program2- Without using a third variable
