```python
Name=input("enter your name:")
Bank_name=input("enter your bank name:")
Balance=5000
print("your bank account balance is:",Balance)
print("n1.withdraw money")
print("n2.deposit money")
choice=int(input("Enter your choice:"))
if choice==1:
  Withdraw=int(input("Enter money to withdraw money:"))
  balance=Balance-Withdraw
  print("your current balance is:",balance)
elif choice==2:
  Deposit=int(input("Enter money to deposit money:"))
  balance=Balance+Deposit
  print("your current balance is:",balance)
else:
  print("insufficient balance:")

print("Thank you",Name,"for using our bank atm")
print("HAVE A WONDERFUL DAY",Name)
```
