```python
year=int(input("Enter a year:"))

if year%4==0:
   print(year,"It is leap year")
elif year%100==0:
   print(year,"It is century year")
elif year%400==0:
   print(year,"It is leap year")
else:
   print(year,"It is normal year")
```

Divisible by 4 ‚Üí Leap Year
Divisible by 400 ‚Üí Leap Year

Divisible by 100   ‚Üí Century Year

Otherwise         ‚Üí Normal year(365 days)
