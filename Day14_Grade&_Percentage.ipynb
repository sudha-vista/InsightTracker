```python

# Day 17 â€“ Grade and Percentage Calculator

print("Enter marks for 5 subjects:")

Eng = float(input(" Enter eng marks: "))
Math = float(input(" Enter math marks: "))
Sci = float(input(" Enter sci marks: "))
Soc = float(input(" Enter soc marks: "))
Cs = float(input(" Enter cs marks: "))

total = Eng+Math+Sci+Soc+Cs
percentage = (total / 500) * 100
percentage=round(percentage,2)

print("total Marks =", total)
print("Percentage =", percentage,"%")


# Grade logic
if percentage>=90:
   print("grade:A")
elif percentage>=80:
   print("grade:B")
elif percentage>=70:
   print("grade:C")
elif percentage>=60:
   print("grade:D")
elif percentage>=50:
   print("grade:E")
else:
   print("grade:fail")
```
