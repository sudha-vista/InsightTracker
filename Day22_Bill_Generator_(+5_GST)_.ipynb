```python
Product1_name=input("Enter product name:")
Product2_name=input("Enter product name:")
Pr1_quantity=int(input("Enter product quantity:"))
Pr2_quantity=int(input("Enter product quantity:"))
Price=int(input("Enter the price for both products:"))
Total_amount=(Price*Pr1_quantity*Pr2_quantity)
Gst=(0.05*Total_amount)
Final_amount=(Total_amount+Gst)

print("Product 1 Name :", Product1_name)
print("Product 2 Name :", Product2_name)
print("Quantity of Product 1 :", Pr1_quantity)
print("Quantity of Product 2 :", Pr2_quantity)
print("Price for both products :", Price)
print("Total Amount :", Total_amount)
print("GST (5%) :", Gst)
print("Final Amount :", Final_amount)
```
