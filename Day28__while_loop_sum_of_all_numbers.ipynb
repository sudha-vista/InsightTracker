```python
N=int(input("Enter how many numbers you want to add:"))
count=0
total=0
while count<N:
   Num=float(input(f"Enter number{count+1}:"))
   total=Num+total
   count=count+1
print("The sum of all no. is:", total)
```

* count = 0 & total = 0
â†’ Start values.
âœ… count helps loop know â€œhow many numbers entered so far.â€
âœ… total stores the sum (starts from 0).







 * while count < N:
â†’ Repeats until count becomes equal to N.
âœ… < not <= because counting starts from 0.







 * Num = float(input(f"Enter number {count + 1}: "))

float() â†’ converts text to decimal number.

f"..." â†’ f-string replaces {count+1} with actual number.
âœ… Example output: â€œEnter number 1:â€, â€œEnter number 2:â€â€¦








* total = total + Num
âœ… Adds the new number to the previous total (running sum).







* count = count + 1
âœ… Increases counter by 1 each loop, or otherwise infinite loop!







* print("The sum of all numbers is:", total)
âœ… Outside loop â†’ shows final answer only once.
