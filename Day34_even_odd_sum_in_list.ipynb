```python
# Step 1: Ask user how many numbers they want to enter
n = int(input("How many numbers do you want to enter? "))


# Step 2: Create an empty list to store numbers
numbers_list = []


# Step 3: Get numbers from the user
for i in range(n):
    num = int(input(f"Enter number {i+1}: "))
    numbers_list.append(num)


# Step 4: Create variables to store sums
even_sum = 0
odd_sum = 0


# Step 5: Loop through the list and calculate sums
for num in numbers_list:
    if num % 2 == 0:
        even_sum += num
    else:
        odd_sum += num


# Step 6: Print the results
print("Sum of even numbers:", even_sum)
print("Sum of odd numbers:", odd_sum)
```

1. i in for i in range(n)

i = loop counter, not entered by user

Counts how many times the loop runs (0 → n−1)

Helps track iteration (e.g., which number is being entered)



---

2. i+1

Used to show friendly numbering to user

Humans count from 1, Python loops from 0

Example: i=0 → i+1=1 → “Enter number 1”



---

3. f"..." (formatted string)

Allows variables inside strings

{i+1} → replaced with actual value of i+1

Example:


i = 0
print(f"Enter number {i+1}")
# Output → Enter number 1
