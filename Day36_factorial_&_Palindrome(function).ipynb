```python
def factorial(num):
  fact=1
  for i in range(1,num+1):
    fact=fact*i
  return fact

num=int(input("Enter number:"))
factorial(num)
print("factorial is:",factorial(num))
```

```python
def is_palindrome(num):
  original=num
  reversed_num=0
  while num>0:
     digit=num%10
     reversed_num=reversed_num*10+digit
     num=num//10
  if reversed_num==original:
     return True
  else:
     return False

num=int(input("Enter number:"))

if is_palindrome(num):
   print("It is palindrome")
else:
   print("It is not palindrome")
```

1️⃣ Function creation

def is_palindrome(num):

Makes a machine called is_palindrome.

num is the number we want to check.



---

2️⃣ Store original number

original = num

Save the original number to compare later.

Because we will reverse num.



---

3️⃣ Start reversed number

reversed_num = 0

This will store the reversed number step by step.



---

4️⃣ Loop to reverse the number

while num > 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num = num // 10

% 10 → gives last digit

Multiply reversed_num by 10 and add digit → builds reversed number

// 10 → removes last digit from num

Repeat until number becomes 0


Example: 123

Step 1: digit = 3 → reversed_num = 0*10+3 = 3 → num = 12

Step 2: digit = 2 → reversed_num = 3*10+2 = 32 → num = 1

Step 3: digit = 1 → reversed_num = 32*10+1 = 321 → num = 0



---

5️⃣ Compare reversed number with original

if reversed_num == original:
    return True
else:
    return False

If same → it’s a palindrome

Else → not a palindrome



---

6️⃣ Take user input

num = int(input("Enter a number: "))

User types number (e.g., 121)

Stored in num



---

7️⃣ Call the function and print result

    print("It is a Palindrome")

    print("It is Not a Palindrome")
