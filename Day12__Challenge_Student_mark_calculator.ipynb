```python
st_name=input("Enter the name of the student:")
st_class=int(input("Enter the class of the student:"))
eng_mark=int(input("Enter the marks of English:"))
math_mark=int(input("Enter the marks of Maths:"))
sci_mark=int(input("Enter the marks of Science:"))
marks_scored=eng_mark+math_mark+sci_mark
percentage=(marks_scored/300)*100
percentage=round(percentage,2)
average_mark=marks_scored/3
average_mark=round(average_mark,2)
print("MARKS SCORED:",marks_scored)
print("PERCENTAGE:",percentage,"%")
print("AVERAGE:",average_mark)
```

**Notes**

* Percentage formula- (got/total)*100

* Average- (sum of all values/number of values)

* average- mean marks per subject

* percentage- score out of 100%

if your subjects had differnt total marks like(50),then average is not equal to percentage.

* For decimal points...use function as:

     round(value,digits)

value: the value you want to round

digits: how many decimal places you want

* example:
round(percent,2)

so percent will be in 2 decimal points.
