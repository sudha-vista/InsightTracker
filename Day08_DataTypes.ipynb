```python
A={1,2,3,4}
B={3,4,5,6}

print("Union:",A|B)     #combines both sets
print("Intersection:",A&B)   #common values
print("Difference:",A-B)  #elements in A but not in B
print("Symmetric Difference:",A^B)   #elements in A or B, not both
```
