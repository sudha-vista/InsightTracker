```python

# Function to find maximum in a list
def find_max(numbers):
    max_num = numbers[0]      # Assume first element is max
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num                  # Return the largest number

# Example list
data = [45, 72, 18, 90, 33]

# Call the function
maximum = find_max(data)

# Display result
print("Maximum number is:", maximum)
```

numbers = parameter (placeholder inside function)
